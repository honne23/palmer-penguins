```python
import time

import pandas as pd
import numpy as np
import typing
import matplotlib.pyplot as plt

from functools import reduce
from sklearn.decomposition import PCA, KernelPCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint


np.random.seed(42)
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
labels = ['sex', 'species']
```

## Create a data-cleaning pipeline
- Data is mean-normalised for use in multiple models

```python
def clean_null_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    Drop all rows that have null values
    """
    print(f'Portion of rows with null values: {round(1-(len(df.dropna())/ len(df)),2)}%')
    return df.dropna()

def collect_subset(df: pd.DataFrame) -> pd.DataFrame:
    """
    Collect a subset of features of interest + labels from data frame
    """
    return df[labels + features]

def normalise(df: pd.DataFrame) -> pd.DataFrame:
    """
    Carry out mean-normalization before creating train-test splits
    """
    df[features] = (df[features] - df[features].mean())/df[features].std()
    return df
```

```python
# Order of data-cleaning transformations indicated by index in list
transformations = [clean_null_values, collect_subset, normalise]

# Transform function applies data transformations successively
transform = lambda x: reduce(lambda d, f : f(d), transformations, x)
```

```python
df = pd.read_csv('data.csv')

# Reproducible transformations
df = transform(df)
```

## Data visualisation of body attributes
- Kernel density estimation is used to estimate the probability density function from their respective histograms
- Most of the features appear to have a bi-model distribution, and yet there are 3 species

```python
def create_subplot(ax, feature:str, df: pd.DataFrame):
    df[feature].plot.kde(ax=ax, legend=False)
    df[feature].plot.hist(density=True, ax=ax)
    ax.set_ylabel('Probability')
    ax.set_xlabel(feature)
    ax.grid(axis='y')
    ax.set_facecolor('#d8dcd6')

def plot_densities(df: pd.DataFrame, by_sex=False): 
    if by_sex:
        fig, axis = plt.subplots(2, len(features))
        fig.set_figheight(8)
        fig.set_figwidth(24)
        sexes = ['male', 'female']
        for sex in range(2):
            for index in range(len(features)) : 
                create_subplot(axis[sex][index], feature=features[index], df= df[df['sex'] == sexes[sex]])
        
    else:
        fig, axis = plt.subplots(1, len(features))
        fig.set_figheight(4)
        fig.set_figwidth(24)
        for index in range(len(features)) :    
            create_subplot(axis[index], feature= features[index], df= df)
```

## Here we can see that most of the attributes are bi-modal without taking 'sex' into account
<h4> Hypothesis:</h4>
Could indicate sexual dimorphism is responsible for the first three features (bill length, bill depth and flipper length)

- However body mass index appears to be 'almost' tri-modal (you can see about 3 bumps)

```python
plot_densities(df, by_sex=False)
```

## Updated hypothesis
- There is indeed sexual dimorphism at play
- But when controlling for gender, there are still multi-modalities
- At least two classes seem linearly separable 
- Thus, the remaining class must be quite similar to one of the others in terms of its moments or at least defined over a subset of the support
- There may be a <b>non-linear</b> combination of features that can further subdivide the similar classes

```python
plot_densities(df, by_sex=True)
```

## Begin to prep data for training with different models

```python
def create_train_test_split(df: pd.DataFrame, test_size: float, features: typing.List[str], include_sex= False):
    """
    Create training and test splits to train classifer models, optionally include sex as a feature
    """
    df_slice = df[['species', *features]]
    if include_sex:
        X = df.drop('species', axis=1)
    else:
        X = df.drop(['species', 'sex'], axis=1) 
    y = df['species']
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=test_size)
    return (X_train, y_train), (X_test, y_test)
```

#### Observations:
- Features 3 & 2 appear to positively co-vary quite strongly (body mass index and flipper length)
- Features 2 & 0 less so (flipper length and bill length)
- Features 2 & 1 appear to negatively co-vary (flipper length and bill_depth)

```python
pca = PCA(n_components = 4)
result = pca.fit(df[features])
pd.DataFrame(result.get_covariance())
```

#### Inspecting the explained variance ratio
- Indicates that bill_length is mostly responsible for the variance in 4 dimensions

```python
pd.DataFrame(result.explained_variance_ratio_)
```

#### Create training and testing splits

```python
training, test = create_train_test_split(df, 0.1, features)
```

#### SVM results confirm non-linearity hypothesis
- Results shown here show here are promising in terms of their information retrieval scores
- After hyperparameter tuning, we seem to find the perfect fit

```python
svc = SVC(kernel='rbf')
svc.fit(training[0], training[1])
y_pred = svc.predict(test[0])
print(classification_report(test[1], y_pred))
```

##### Tune hyper-parms

```python
svc = SVC(kernel='rbf')
distributions = dict(kernel=['linear', 'rbf'], C=uniform(0.1, 1000), gamma=uniform(0.1,5))
clf = RandomizedSearchCV(svc, distributions, random_state=0)

start = time.time()
clf.fit(training[0], training[1])
end = time.time()

print(f'Elapsed time in seconds: {round(end - start, 5)}s\n')

y_pred = clf.predict(test[0])

print(classification_report(test[1], y_pred))
```

```python
plot_confusion_matrix(clf, test[0], test[1],display_labels=np.sort(df['species'].unique()),cmap=plt.cm.Blues,normalize='true')
```

#### Test a k-means classifier as a sanity-check
- Maybe the SVM overcomplicates the task
- k-means appears to perform quite well when combined with logistic regression, although not as well as the SVM
- We can see the Adelie and Chinstrap clusters appear to be more difficult to distinguish
- Further validates the presence of potential non-linear relationships between features

```python
pipeline = Pipeline([
    ('kmeans', KMeans(n_clusters=3, random_state=0)),
    ('lreg', LogisticRegression())
])
clf = pipeline.fit(training[0], training[1])

y_pred = clf.predict(test[0])

print(classification_report(test[1], y_pred))

plot_confusion_matrix(clf, test[0], test[1],display_labels=np.sort(df['species'].unique()),cmap=plt.cm.Blues,normalize='true')
```

#### Try to distill the features to make k-means perform better
- Here I'm attempting to tease out the non-linearity by first applying Kernel PCA
- This is to preserve that variance under non-linear constraints when projecting to lower dimensions
- It does not improve the perfomance of the dataset
- Indicates that cluster-wise similarity is not an optimal metric for this dataset

```python
distributions = [{
        "kpca__gamma": uniform(0.03, 10),
        "kpca__kernel": ["rbf", "sigmoid"]
    }]
pipeline = Pipeline([
    ('kpca', KernelPCA(n_components = 2, kernel='rbf')),
    ('kmeans', KMeans(n_clusters=3, random_state=0)),
    ('lreg', LogisticRegression())
])
clf = RandomizedSearchCV(pipeline, distributions, random_state=0).fit(training[0], training[1])
y_pred = clf.predict(test[0])

print(classification_report(test[1], y_pred))

plot_confusion_matrix(clf, test[0], test[1],display_labels=np.sort(df['species'].unique()),cmap=plt.cm.Blues,normalize='true')
```

#### Gradient boosting is as another good candidate for classification
- Dataset is relatively small
- Ensemble methods may be more effective
- Appears to be extremely effective, achieving scores of 1.0 with some default parameters

```python
gBoost = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0)
gBoost.fit(training[0], training[1])

y_pred = gBoost.predict(test[0])

print(classification_report(test[1], y_pred))
```

#### Hyperparameter optimization significantly increases training time as compared with the SVM
- Due to depth and estimator hyper-parameters
- Still performs very well on the dataset
- Confusion matrix confirms the similarities between the distributions of the Chinstrap and Adelie species
- Some rounds of optimization produce a final model that performs worse than others
- Given computational time, and consistency, optimized SVM appears to be the best approach.

```python
gBoost = GradientBoostingClassifier()
distributions = dict(n_estimators=randint(20, 300), learning_rate = uniform(0.1, 10), max_depth=randint(1, 3) )
clf = RandomizedSearchCV(gBoost, distributions, random_state=0)

start = time.time()
clf.fit(training[0], training[1])
end = time.time()

print(f'Elapsed time in seconds: {round(end - start, 5)}s\n')

y_pred = clf.predict(test[0])

print(classification_report(test[1], y_pred))
```

```python
plot_confusion_matrix(clf, test[0], test[1],display_labels=np.sort(df['species'].unique()),cmap=plt.cm.Blues,normalize='true')
```
